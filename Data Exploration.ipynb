```python
import pandas as pd
data=pd.read_excel('MIAS.xlsx')

# data.iloc[:, 2].hist(figsize=(5,5),edgecolor='black',color='red')
data.head
```

```python
data.iloc[:, 2].hist(figsize=(5,5),edgecolor='black', color='red');
```

```python
from matplotlib import pyplot as plt
plt.rcParams.update({'font.family':'sans-serif'})
plt.rcParams.update({'font.size':8})
data.iloc[:, 2].value_counts().plot.pie(counterclock=False, shadow=True)
plt.title('MIAS Mammogram Categories')

plt.ylabel('Class')
```

```python
import seaborn as sns
sns.countplot(x=data.iloc[:, 3], hue=data.iloc[:, 3], data=data)
plt.legend(['Benign','Malignant'])
plt.title('Severity of Abnormality')
plt.xlabel('Severity')
plt.ylabel('Number of Cases')
plt.rcParams.update({'font.family':'sans-serif'})
plt.rcParams.update({'font.size':8})
```

```python
plt.scatter(data.iloc[:, 4], data.iloc[:, 5])

plt.title('Scatter Plot')
plt.ylabel('Y coordinates of centre of abnormality')
plt.xlabel('X coordinates of centre of abnormality')
plt.show()
```
